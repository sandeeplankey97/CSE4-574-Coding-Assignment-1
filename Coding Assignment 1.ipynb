```python
# Selecting features
selected_features = ['fixed acidity', 'volatile acidity','residual sugar' ,'citric acid', 'chlorides', 
                     'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'quality']

# Extracting features and target variable
X = data[selected_features].values
y = data['alcohol'].values

# Adding a bias term to X for the intercept
X = np.c_[np.ones(X.shape[0]), X]
```

```python
# Splitting the data into training and test sets (80%-training, 20%-testing)
train_size = int(0.8 * len(data))

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
print('********* Shapes for Traing and Testing data *********')
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Calculating the weights using closed-form solution
weight = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
print(weight)
```

```python
# Calculating predictions and mse on the test data
y_pred = X_test @ weight

m_sq_er = np.mean((y_test - y_pred) ** 2)
print("Mean Squared Error :", m_sq_er)
```

```python
print('********* Plotting graph for Predictions and Actual Data *********')

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='violet', alpha=0.7, edgecolors='black')
plt.title('Relation between Predictions and Actual Alcohol', fontsize=16)
plt.xlabel('Actual Alcohol', fontsize=14)
plt.ylabel('Predicted Alcohol', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Reading Penguins data
pen_data = pd.read_csv('/Users/sandeeplankey/Documents/Sem 2/ML/Coding Assignment/datasets/penguins.csv')
```

```python
print('********* Penguins data head *********')
print(pen_data.head())
```

```python
print('********* Checking the missing Data *********')
print(pen_data.isnull().sum())
```

```python
# Replacing missing values in data
print('********* Data after replacing the missing Data *********')
#pen_data = pen_data.fillna(0)
pen_data = pen_data.dropna()
print(pen_data.isnull().sum())
```

```python
pen_data.head()
```

```python
# Converting features with string datatype to categorical
pen_data['species'] = pen_data['species'].astype('category').cat.codes
pen_data['island'] = pen_data['island'].astype('category').cat.codes
pen_data['sex'] = pen_data['sex'].astype('category').cat.codes

#norm_data = (pen_data - pen_data.min()) / (pen_data.max() - pen_data.min())
#X = norm_data[['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

# Normalizing non-categorical features
pen_data['bill_length_mm'] = (pen_data['bill_length_mm'] - pen_data['bill_length_mm'].min()) / (pen_data['bill_length_mm'].max() - pen_data['bill_length_mm'].min())
pen_data['bill_depth_mm'] = (pen_data['bill_depth_mm'] - pen_data['bill_depth_mm'].min()) / (pen_data['bill_depth_mm'].max() - pen_data['bill_depth_mm'].min())
pen_data['flipper_length_mm'] = (pen_data['flipper_length_mm'] - pen_data['flipper_length_mm'].min()) / (pen_data['flipper_length_mm'].max() - pen_data['flipper_length_mm'].min())
pen_data['body_mass_g'] = (pen_data['body_mass_g'] - pen_data['body_mass_g'].min()) / (pen_data['body_mass_g'].max() - pen_data['body_mass_g'].min())

# Splitting the data into features (X) and target variable (y)
X = pen_data[['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y = pen_data['sex']
```

```python
class LogitRegression:
    def __init__(self, learn_rate, iters):
        self.learn_rate = learn_rate
        self.iters = iters

    def sigmoid(self, lin_comb):
        return 1 / (1 + np.exp(-lin_comb))

    def cost(self, pred_prob, y):
        return (-y * np.log(pred_prob) - (1 - y) * np.log(1 - pred_prob)).mean()

    def gradient_descent(self, X, pred_prob, y):
        return np.dot(X.T, (pred_prob - y)) / y.shape[0]

    def fit(self, X, y):
        self.w_v = np.zeros(X.shape[1])
        self.losses = []

        for ele in range(self.iters):
            lin_comb = np.dot(X, self.w_v)
            pred_prob = self.sigmoid(lin_comb)
            gradt = self.gradient_descent(X, pred_prob, y)
            self.w_v -= self.learn_rate * gradt
            loss = self.cost(pred_prob, y)
            self.losses.append(loss)

    def predict(self, X):
        print(self.w_v)
        return np.round(self.sigmoid(np.dot(X, self.w_v)))
```

```python
# Splitting the data into training and test sets (80%-training, 20%-testing)
train_dat_size = int(0.8 * len(X))
X_train, X_test = X[:train_dat_size], X[train_dat_size:]
y_train, y_test = y[:train_dat_size], y[train_dat_size:]
```

```python
# Initializing and train the model
# 1st hyperparameter tuning
learn_rate = 1e-3
iters = 100000
model = LogitRegression(learn_rate, iters)
model.fit(X_train, y_train)

# Making predictions on the test set
predns = model.predict(X_test)

# Calculating accuracy
acc = np.mean(predns == y_test)
print("Accuracy: {:}%".format(acc * 100))

plt.figure(figsize=(10, 6))
plt.plot(range(iters), model.losses, color='violet', linewidth=2, marker='o', markersize=5, label='Training Loss')
plt.title('Relation between Loss and Iterations', fontsize=16)
plt.xlabel('Number of Iterations', fontsize=14)
plt.ylabel('Loss percentage (%)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2nd hyperparameter tuning
learn_rate = 1e-4
iters = 115000
model = LogitRegression(learn_rate, iters)
model.fit(X_train, y_train)

# Making predictions on the test set
predns = model.predict(X_test)

# Calculating accuracy
acc = np.mean(predns == y_test)
print("Accuracy: {:}%".format(acc * 100))

plt.figure(figsize=(10, 6))
plt.plot(range(iters), model.losses, color='violet', linewidth=2, marker='o', markersize=5, label='Training Loss')
plt.title('Relation between Loss and Iterations', fontsize=16)
plt.xlabel('Number of Iterations', fontsize=14)
plt.ylabel('Loss percentage (%)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
